```python
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"
t_dataframe = pd.read_csv(url)
t_dataframe
```

```python
#Find how many duplicate rows exist in the dataframe.
t_dataframe[t_dataframe.duplicated()==True].shape[0]
```

```python
#Remove the duplicate rows from the dataframe
t_dataframe.drop_duplicates(inplace=True)
```

```python
#Verify if duplicates were actually dropped
t_dataframe[t_dataframe.duplicated()==True]
```

```python
#Find the missing values for all columns
t_dataframe.isnull().any()
```

```python
#Find out how many rows are missing values in the column 'WorkLoc'
t_dataframe[t_dataframe['WorkLoc'].isnull()==True].shape[0]
```

```python
#Find the  value counts for the column WorkLoc.
t_result = t_dataframe[t_dataframe['WorkLoc'].isnull()==False]
t_result['WorkLoc'].unique()
for values in t_result['WorkLoc'].unique():
   print(values)
```

```python
#Identify the value that is most frequent (majority) in the WorkLoc column.
majority = t_dataframe.groupby(by=['WorkLoc'])['WorkLoc'].count().sort_values(ascending=False).head(1)
majority
```

```python
# Imputing or replace all the empty rows in the column WorkLoc with the value that you have identified as majority.
t_dataframe['WorkLoc'].fillna(majority.index[0], inplace=True)
```

```python
#Verify if imputing was successful
t_dataframe[t_dataframe['WorkLoc'].isnull()==True].shape[0]
#t_dataframe.groupby(by=['WorkLoc'])['WorkLoc'].count().sort_values(ascending=False).head(1)
```

```python
#Normalize data
#List out the various categories in the column 'CompFreq'
t_dataframe['CompFreq'].fillna("", inplace=True)
for items in t_dataframe['CompFreq'].unique():
    if bool(items):
        print(items)
```

```python
#Create a new column named 'NormalizedAnnualCompensation' based on available data
#ConvertedComp: Salary converted to annual USD salaries using the exchange rate on 2019-02-01, 
#assuming 12 working months and 50 working weeks.
for index, rows in t_dataframe.iterrows():
    if rows['CompFreq'] == 'Yearly':
        t_dataframe.loc[index,'NormalizedAnnualCompensation'] = rows['CompTotal']
    if rows['CompFreq']=='Monthly':
        t_dataframe.loc[index,'NormalizedAnnualCompensation'] = rows['CompTotal']*12
    if rows['CompFreq']=='Weekly':
        t_dataframe.loc[index,'NormalizedAnnualCompensation'] = rows['CompTotal']*50
        
t_dataframe[['Respondent','CompFreq','CompTotal','NormalizedAnnualCompensation','CurrencySymbol','CurrencyDesc']].sort_values(by=['Respondent'])
    
```
